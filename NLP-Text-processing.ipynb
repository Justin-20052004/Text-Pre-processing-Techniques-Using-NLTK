```python
import nltk
import string
import numpy as np

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import pos_tag

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.cluster import KMeans

from gensim.models import Word2Vec
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
text = """
Artificial intelligence (AI) is transforming the modern world: it helps businesses analyze data,
supports doctors in healthcare, and improves everyday technology. Many people ask, “How does AI actually work?” 
Well, AI systems learn from large amounts of information, recognize patterns, and respond to user requests! For example, 
voice assistants like Siri, Alexa, or Google Assistant listen to commands and provide helpful answers. In programming and documentation,
developers sometimes organize information using structures such as {system settings} or lists like [input, process, output]. Isn’t it fascinating?

In the future, AI may influence many areas: education, healthcare, transportation, and entertainment. 
Imagine this: smart cars driving automatically, robots helping in hospitals, and personalized lessons for students! 
Some experts describe system components using formats like configuration{language: "English"} or task lists [planning, development, testing]. 
People still ask, “Will AI replace human jobs?” The answer isn’t simple; it depends on how technology is designed, used, and managed. Ultimately,
with careful development and responsible use, AI can benefit society and improve human life.
"""
```

```python
tokens = word_tokenize(text.lower())

tokens = [word for word in tokens if word.isalnum()]

print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word not in stop_words]

print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("Stemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemma_without_pos = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Lemmatization without POS:\n", lemma_without_pos)
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

pos_tags = pos_tag(filtered_words)

lemma_with_pos = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

print("Lemmatization with POS:\n", lemma_with_pos)
```

```python
processed_text = " ".join(filtered_words)

# BoW
bow = CountVectorizer()
bow_matrix = bow.fit_transform([processed_text])

print("BoW Features:\n", bow.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())

# TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([processed_text])

print("\nTF-IDF Features:\n", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
tokens2 = word_tokenize(text.lower())
tokens2 = [word for word in tokens2 if word.isalnum()]

data = [tokens2]

model_w2v = Word2Vec(
    sentences=data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=100
)

print(model_w2v.wv.index_to_key)
```

```python
def safe_similarity(w1, w2):
    if w1 in model_w2v.wv and w2 in model_w2v.wv:
        return model_w2v.wv.similarity(w1, w2)
    else:
        return f"{w1} or {w2} not in vocabulary"

print("AI vs data:", safe_similarity('ai', 'data'))
print("learning vs data:", safe_similarity('learning', 'data'))
print("doctors vs healthcare:", safe_similarity('doctors', 'healthcare'))
```

```python
sentences = [
    "Artificial intelligence is transforming the world",
    "AI is changing industries and technology",
    "Children are playing football in the park"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

similarity = cosine_similarity(X)

print("Similarity Matrix:\n", similarity)

print("S1 vs S2:", similarity[0][1])
print("S1 vs S3:", similarity[0][2])
print("S2 vs S3:", similarity[1][2])
```

```python
texts = [
    "AI is transforming industries",
    "Machine learning improves healthcare",
    "Deep learning is powerful",
    "Football is a popular sport",
    "Cricket is played worldwide",
    "Tennis is an outdoor game"
]

labels = ["tech", "tech", "tech", "sports", "sports", "sports"]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
test_sentence = ["AI improves technology"]

test_vec = vectorizer.transform(test_sentence)

prediction = clf.predict(test_vec)

print("Prediction:", prediction[0])
print("Accuracy:", clf.score(X_test, y_test))
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

clusters = kmeans.labels_

print("Cluster Labels:\n", clusters)

for i in range(len(texts)):
    print(texts[i], "→ Cluster", clusters[i])
```
