```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
text = """
Artificial intelligence (AI) is transforming the modern world: it helps businesses analyze data,
supports doctors in healthcare, and improves everyday technology. Many people ask, “How does AI actually work?” 
Well, AI systems learn from large amounts of information, recognize patterns, and respond to user requests! For example, 
voice assistants like Siri, Alexa, or Google Assistant listen to commands and provide helpful answers. In programming and documentation,
developers sometimes organize information using structures such as {system settings} or lists like [input, process, output]. Isn’t it fascinating?

In the future, AI may influence many areas: education, healthcare, transportation, and entertainment. 
Imagine this: smart cars driving automatically, robots helping in hospitals, and personalized lessons for students! 
Some experts describe system components using formats like configuration{language: "English"} or task lists [planning, development, testing]. 
People still ask, “Will AI replace human jobs?” The answer isn’t simple; it depends on how technology is designed, used, and managed. Ultimately,
with careful development and responsible use, AI can benefit society and improve human life.
"""

print("Original Text:\n", text)
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
tokens_no_punct = [word for word in tokens if word.isalnum()]
print(tokens_no_punct)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens_no_punct if word.lower() not in stop_words]

print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("Stemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemma_without_pos = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Lemmatization without POS:\n", lemma_without_pos)
```

```python
def get_wordnet_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
pos_tags = pos_tag(filtered_words)

lemma_with_pos = []

for word, tag in pos_tags:
    pos = get_wordnet_pos(tag)
    lemma_with_pos.append(lemmatizer.lemmatize(word, pos))

print("Lemmatization with POS:\n", lemma_with_pos)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# Initialize CountVectorizer
bow = CountVectorizer()

# Fit and transform
bow_matrix = bow.fit_transform([text])

# Convert to array
bow_array = bow_matrix.toarray()

# Get feature names
features = bow.get_feature_names_out()

print("BoW Features:\n", features)
print("\nBoW Matrix:\n", bow_array)
```

```python
tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform([text])

tfidf_array = tfidf_matrix.toarray()

features = tfidf.get_feature_names_out()

print("TF-IDF Features:\n", features)
print("\nTF-IDF Matrix:\n", tfidf_array)
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
```

```python
tokens = word_tokenize(text.lower())

# Convert into sentences (Word2Vec needs list of lists)
data = [tokens]
```

```python
model = Word2Vec(
    sentences=data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    epochs=100
)
```

```python
print(model.wv.index_to_key)
```

```python
print("Similarity between AI and technology:",
      model.wv.similarity('ai', 'technology'))

print("Similarity between data and information:",
      model.wv.similarity('data', 'information'))

print("Similarity between doctor and hospitals:",
      model.wv.similarity('doctors', 'hospitals'))

print("Similarity between influence and example:",
      model.wv.similarity('influence', 'example'))
```

```python
print(model.wv.most_similar('ai'))
```
